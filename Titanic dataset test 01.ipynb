```python
df['age'].plot(kind="box")
```

```python
q1=df['fare'].quantile(0.25)
q3=df['fare'].quantile(0.75)
```

```python
iqr=q3-q1
```

```python
upper_bound=q3+(1.5*iqr)
lower_bound=q1-(1.5*iqr)
```

```python
fare_arr=df['fare']
lower=np.percentile(fare_arr,1)
upper=np.percentile(fare_arr,99)

print("Lower is :",lower)
print("Upper is :",upper)
```

```python
for i in df['fare']:
    if i <lower_bound:
        df['fare']=df['fare'].replace(i,lower)
        
    elif i> upper_bound:
        df['fare']=df['fare'].replace(i,upper)
df        
```
