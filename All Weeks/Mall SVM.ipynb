```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler
```

```python
# Read csv file

df=pd.read_csv("C:/Users/Pramoda A S/Desktop/AIML Documents/DataSets in CSV files/Mall_Customers.csv")
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler

X=df.drop(columns=['CustomerID','Gender'])
Y=df['Gender']
```

```python
X
```

```python

lf=StandardScaler()
x=lf.fit_transform(X)
```

```python
x
```

```python
lk=LabelEncoder()
y=lk.fit_transform(Y)
```

```python
y
```

```python
# Split into Train/Test
X_train, X_test, y_train, y_test = train_test_split(x,y ,test_size=0.3,random_state=42,stratify=y)
```

```python
# -----------------------------
# 3. Logistic Regression Model
# -----------------------------
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# Predictions
y_pred = log_reg.predict(X_test)
y_prob = log_reg.predict_proba(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
# -----------------------------
# 4. Evaluation
# -----------------------------
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
labels = sorted(list(set(y_test)))
print("\nConfusion Matrix:\n")
print(metrics.confusion_matrix(y_test, y_pred, labels=labels))

```

```python
from sklearn import metrics

# Get unique labels
labels = sorted(list(set(y_test)))

# Generate report
print(metrics.classification_report(
    y_test, y_pred, labels=labels, target_names=[str(x) for x in labels], zero_division=0
))
```

```python
from sklearn.svm import SVC

# -----------------------------
# 3. SVM
# -----------------------------
log_reg = SVC()
log_reg.fit(X_train, y_train)

# Predictions
y_pred = log_reg.predict(X_test)
#y_prob = log_reg.predict_proba(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
# -----------------------------
# 4. Evaluation
# -----------------------------
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
labels = sorted(list(set(y_test)))
print("\nConfusion Matrix:\n")
print(metrics.confusion_matrix(y_test, y_pred, labels=labels))

```
