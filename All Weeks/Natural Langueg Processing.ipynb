## Word Tokanization

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
stwd=set(STOPWORDS)

wordcloud=WordCloud(width=800, height=800,background_color='white', stopwords=stwd,min_font_size=10).generate(text)

# Plot the wordcloud 

plt.figure(figsize=(4,4),facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
df=open("C:/Users/Pramoda A S/Desktop/AIML Documents/nlpEx.txt")
text=df.read()
print(text)
```

```python
nltk.tokenize.word_tokenize(text,language='english',preserve_line=False)
```

```python
from nltk.tokenize import word_tokenize

words=word_tokenize(text)
```

```python
print("Total number of words :",len(words))
print(words)
```

```python
from nltk.probability import FreqDist

all_fdist=FreqDist(words)
all_fdist['performance']
```

```python
import matplotlib.pyplot as plt 
import pandas as pd 

all_fdist=pd.Series(dict(all_fdist))

fig, ax=plt.subplots(figsize=(15,15))

all_fdist.plot(kind='bar')
plt.title("Frequency Disrtubution of words")
plt.ylabel("Counts")
plt.show()
```
