# Consider the Student dataset and train simple linear regression to predict the student score based on student houres and also evaluate the model using different metrics.

```python
# Import the libraries

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#Read CSV File

df=pd.read_csv("C:/Users/Pramoda A S/Desktop/AIML Documents/DataSets in CSV files/student_scores.csv")
df.head(5)
```

```python
# Show the numbers of rows and columns

df.shape
```

```python
# Pre-processing 
# show the missing values

df.isnull().sum()
```

```python
# Fill the missing values using mean

a=df['Scores'].mean()
a
```

```python
df['Scores']=df['Scores'].fillna(a)
```

```python
df.isnull().sum()
```

```python
# Splitting Data

from sklearn.linear_model import LinearRegression

x=df[['Hours']]
y=df['Scores']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=22)
```

```python
# Fit the model

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
# Predict the model

y_pred = model.predict(x_test)
y_pred
```

```python
# Evaluate the model using MSE and r2_score

print("The MSE is : ", mean_squared_error(y_test,y_pred))
print("The R square is : ", r2_score(y_test, y_pred))
```
