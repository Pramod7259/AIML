# Univarite analysis

```python
# Import the Libraries

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# read csv file

df=pd.read_csv('C:/Users/Pramoda A S/Desktop/AIML Documents/DataSets in CSV files/auto-mpg.csv')
df
```

## Using univariate ploting Histogram 

```python
plt.hist(df['cylinders'])
plt.title('Counting cylinders')
plt.xlabel("Cylinders")
plt.ylabel('Cylinders counts')
plt.show()
```

## Using univariate ploting Line graph

```python
# 1)

df['model year'].value_counts().sort_index().plot(kind="line",marker='o')
```

```python
# 2)

df['cylinders'].value_counts().sort_index().plot(kind='line',marker='o')
```

## Using univariate ploting Box Plot

```python
#  1) 

df['mpg'].plot(kind='box')
```

```python
# 2)

plt.boxplot(df['weight'])
plt.show()
```

 ## Using univariate ploting Bar graph

```python
df['cylinders'].value_counts().plot(kind='bar')
```

## Using univariate ploting Pie chart

```python
df['cylinders'].value_counts().sort_index().plot(kind='pie',autopct='%1.2f%%')
plt.legend()
plt.show()
```

## Using univariate ploting Scatter plot

```python
# Scatter plot

k=df['model year'].value_counts()
plt.scatter(k.index,k.values)
```

# Bivarite analysis

## Using titanic dataset perform the bivariate analysis.

```python
# import libraris

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# read csv file

df=pd.read_csv("C:/Users/Pramoda A S/Desktop/AIML Documents/DataSets in CSV files/titanic.csv")
df
```

```python
# Using the bivariate analyis plot line graph

a=df.groupby(['survived','sex']).size().unstack()
a.plot(kind='line',marker='o')
plt.title("survived VS sex")
plt.show()
```

```python
# Using the bivariate analyis plot Bar graph

b=df.groupby(['survived','sex']).size().unstack()
b.plot(kind='bar')
plt.title("survived VS sex")
plt.show()
```

```python
# Using the bivariate analyis plot Scatter graph

plt.scatter(df['age'],df['fare'])
plt.title("Fare vs Age")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
