```python
q1=df2['age'].quantile(0.25)
q3=df2['age'].quantile(0.75)
```

```python
up_bound=q3+(1.5*iqr)
low_bound=q1-(1.5*iqr)
```

```python
a_arr=df2['age']
out=a_arr[(a_arr<=low_bound) | (a_arr>=up_bound)]
print(out)
```

```python
age_arr=df2['age']
low_cap=np.percentile(age_arr,1)
up_cap=np.percentile(age_arr,99)
```

```python
for i in df2['age']:
    if i<low_bound:
        df2['age']=df2['age'].replace(i,low_cap)
    elif i>up_bound:
        df2['age']=df2['age'].replace(i,up_cap)
```
